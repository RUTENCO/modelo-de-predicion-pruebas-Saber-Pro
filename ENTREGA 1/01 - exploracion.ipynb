```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ruta donde se encuentra el archivo descargado desde Kaggle
file_path = '/content/drive/MyDrive/train.csv'
# Cargar el archivo CSV en un DataFrame de Pandas
df = pd.read_csv(file_path)
# copia del dataframe
dl= df
```

```python
# resumen estadístico de las columnas numéricas en un DataFrame
description = dl.describe()
description
```

```python
# Filtrar los IDs de las personas que tienen internet (FAMI_TIENEINTERNET == 'si'),primeras y ultimas filas porque son muchos datos
ids_con_internet = df.loc[df['FAMI_TIENEINTERNET'] == 'Si', 'ID']
print("\nIDs de las personas que tienen internet:\n")
print(ids_con_internet)
```

```python
# Contar el número de estudiantes por programa academico, da una musetra general porque son muchos datos
print("\nNúmero de estudiantes por clase:\n")
n_estudent_class= dl['ESTU_PRGM_ACADEMICO'].value_counts()
n_estudent_class
```

```python
# Imprimir el número de estudiantes por programa academico de forma completa
n_estudent_class= dl['ESTU_PRGM_ACADEMICO'].value_counts()
print("\nNúmero de estudiantes por clase:\n")
for index, value in n_estudent_class.items():
    print(f"{index}: {value}")
```

```python
# Contar el número de estudiantes por localidad
print("\nNúmero de estudiantes por localidad:\n")
n_estudent_location = dl['ESTU_PRGM_DEPARTAMENTO'].value_counts()
n_estudent_location
```

```python
# Histograma de una variable numerica "periodo"
plt.hist(df['PERIODO'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel('PERIODO')
plt.ylabel('Frecuencia')
plt.title('Histograma de periodos')
plt.show()
```

```python
# gráfico de barras, frecuencia o el recuento de cada categoría en una variable cualitativa utilizando barras.

# Obtener la frecuencia de valores en la columna 'ESTU_PRGM_DEPARTAMENTO'
frecuencia = df['ESTU_PRGM_DEPARTAMENTO'].value_counts()
frecuencia.plot.bar()
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.title('Gráfico de Barras de programa por departamento')
plt.show()
```
