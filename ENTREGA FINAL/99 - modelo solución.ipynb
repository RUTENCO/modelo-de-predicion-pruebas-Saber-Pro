```python
# Eliminar columnas que son despreciables para el modelo

def columns_clean(data):

  # Eliminar la columna "FAMI_TIENEINTERNET.1" del DataFrame original ya que es una columna que repite datos
  data = data.drop(columns=['FAMI_TIENEINTERNET.1'])
  # Eliminar la columna "ESTU_PRIVADO_LIBERTAD" del DataFrame original ya que es una columna que no es muy significativa
  data = data.drop(columns=['ESTU_PRIVADO_LIBERTAD'])
  # Eliminar la columna "ESTU_PRIVADO_LIBERTAD" del DataFrame original ya que es una columna que no es muy significativa
  data = data.drop(columns=['ID'])

  return data
```

```python
# Quitar acentos y poner todo en minusculas

def unidecode_ESTU_PRGM_ACADEMICO(data):

  # Aplicar unidecode() y convertir a minúsculas solo para la columna 'ESTU_PRGM_ACADEMICO'
  data['ESTU_PRGM_ACADEMICO'] = data['ESTU_PRGM_ACADEMICO'].apply(lambda x: unidecode(str(x).lower()))

  return data
```

```python
# se generaliza un poco mas las carreras para facilitar la limpieza de datos

def generalization_carrer(data):

  # Reemplazar "en" por "de"
  data['ESTU_PRGM_ACADEMICO'] = data['ESTU_PRGM_ACADEMICO'].str.replace(r'\b(en)\b', 'de', regex=True)

  # Reemplazar "-" por "y"
  data['ESTU_PRGM_ACADEMICO'] = data['ESTU_PRGM_ACADEMICO'].str.replace('-', 'y')

  # Reemplazar "del" por "de"
  data['ESTU_PRGM_ACADEMICO'] = data['ESTU_PRGM_ACADEMICO'].str.replace(r'\b(del)\b', 'de', regex=True)

  # Reemplazar las comas por un espacio en blanco
  data['ESTU_PRGM_ACADEMICO'] = data['ESTU_PRGM_ACADEMICO'].str.replace(',', '', regex=True)

  # Reemplazar los puntos por un espacio en blanco
  data['ESTU_PRGM_ACADEMICO'] = data['ESTU_PRGM_ACADEMICO'].str.replace(':', '', regex=True)

  # Reemplazar múltiples espacios por un solo espacio
  data['ESTU_PRGM_ACADEMICO'] = data['ESTU_PRGM_ACADEMICO'].str.replace(r'\s+', ' ', regex=True)

  return data
```

```python
# eliminar carreras del data set de entrenamiento que no estan en el data set de test

def delete_categories(data):
  # Lista de valores únicos que deseas eliminar
  valores_a_eliminar = [
      'animacion', 'formacion profesional de deporte', 'gestion de sistemas de informacion',
      'ingenieria de diseno de maquinas y productos industriales', 'ingenieria de logistica y operaciones',
      'licenciatura de ciencias de la educacion', 'licenciatura de educacion basica con enfasis de el area de tecnologia e informatica',
      'licenciatura de educacion basica primaria enfasis de ciencias sociales', 'licenciatura de educacion enfasis de informatica',
      'licenciatura de educacion fisica recreacion y deportes enfasis rural', 'licenciatura de educacion religiosa con enfasis de biblia',
      'licenciatura de espanol y filologia', 'licenciatura de pedagogia reeducativa', 'mercadeo agroindustrial',
      'profesional de artes gastronomicas', 'profesional de investigacion criminal', 'profesional universitario de diseno textil para la industria',
      'quimica y farmacia'
  ]

  # Filtrar el DataFrame eliminando las filas que contienen los valores únicos
  data = data[~data['ESTU_PRGM_ACADEMICO'].isin(valores_a_eliminar)]

  return data
```

```python
# se agregan carreras del data set de test que no estan en el data set de entrenamiento

def add_carrer_dtr(dtr,dts):
  # Valores únicos en dts['ESTU_PRGM_ACADEMICO'] que no están en dtr['ESTU_PRGM_ACADEMICO']
  valores_unicos_dts = dts[~dts['ESTU_PRGM_ACADEMICO'].isin(dtr['ESTU_PRGM_ACADEMICO'])]

  # Agregar las filas de valores_unicos_dts a dtr
  dtr = pd.concat([dtr, valores_unicos_dts], ignore_index=True)

  return dtr
```

```python
# LLenar  todas las filas que contienen valores NaN con la moda

def fill_nan_with_moda(data):

  data = data.fillna(data.mode().iloc[0])

  return data
```

```python
# One hot para 'ESTU_PRGM_ACADEMICO' variable categorica

def one_hot_ESTU_PRGM_ACADEMICO(data):

  # Crear variables ficticias (one-hot encoding) para la columna de lugar
  place_dummies = pd.get_dummies(data['ESTU_PRGM_ACADEMICO'], dtype=int)
  # Eliminar la columna original de lugar del DataFrame
  data.drop('ESTU_PRGM_ACADEMICO', axis=1, inplace=True)
  # Concatenar el DataFrame original con las nuevas columnas de variables ficticias
  data = pd.concat([data, place_dummies], axis=1)

  return data
```

```python
# funcion que aplicara el preprocesado de datos a ambos datasets ademas de agregar al train las nuevas carreras tratadas en el tets

def clean_datas(dtr,dts):

  dtr = columns_clean(dtr)
  dtr = unidecode_ESTU_PRGM_ACADEMICO(dtr)
  dtr = generalization_carrer(dtr)
  dtr = clean_special_character(dtr)
  dtr = generalize_with_regular_expression(dtr)
  dtr = group_carrer(dtr)
  dtr = group_carrer_plus(dtr)
  dtr = delete_categories(dtr)

  dts = columns_clean(dts)
  dts = unidecode_ESTU_PRGM_ACADEMICO(dts)
  dts = generalization_carrer(dts)
  dts = clean_special_character(dts)
  dts = generalize_with_regular_expression(dts)
  dts = group_carrer(dts)
  dts = group_carrer_plus(dts)
  dtr = add_carrer_dtr(dtr,dts)

  dtr = fill_nan_with_moda(dtr)
  dtr = one_hot_ESTU_PRGM_ACADEMICO(dtr)
  dtr = one_hot_ESTU_PRGM_DEPARTAMENTO(dtr)
  dtr = coding_ESTU_VALORMATRICULAUNIVERSIDAD(dtr)
  dtr = coding_ESTU_HORASSEMANATRABAJA(dtr)
  dtr = coding_FAMI_ESTRATOVIVIENDA(dtr)
  dtr = coding_yes_not(dtr)
  dtr = coding_FAMI_EDUCACIONPADRE(dtr)
  dtr = coding_FAMI_EDUCACIONMADRE(dtr)
  dtr = coding_RENDIMIENTO_GLOBAL(dtr)

  dts = fill_nan_with_moda(dts)
  dts = one_hot_ESTU_PRGM_ACADEMICO(dts)
  dts = one_hot_ESTU_PRGM_DEPARTAMENTO(dts)
  dts = coding_ESTU_VALORMATRICULAUNIVERSIDAD(dts)
  dts = coding_ESTU_HORASSEMANATRABAJA(dts)
  dts = coding_FAMI_ESTRATOVIVIENDA(dts)
  dts = coding_yes_not(dts)
  dts = coding_FAMI_EDUCACIONPADRE(dts)
  dts = coding_FAMI_EDUCACIONMADRE(dts)

  return dtr,dts
```

# apartir de varias pruebas, busquedad de informacion, combinacion de hiperparametros. Se logra crear una red neuronal con tensorflow.

```python
# primero iniciamos con Estandarizar las características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Definir el modelo para la red neuronal
def create_model():
    model = Sequential()
    model.add(Dense(512, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    model.add(Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    model.add(Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    model.add(Dense(4, activation='softmax'))

    optimizer = Adam(learning_rate=0.0001)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model
```

```python
# se  Crear el modelo de red neuronla
model = create_model()

# Define callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5, min_lr=1e-6, verbose=1)
```

```python
# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=100, batch_size=128, validation_split=0.2, callbacks=[early_stopping, reduce_lr])
```

```python
# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```
